```python
import numpy as np
import matplotlib.pyplot as plt
# Data structure: [Restaurant_id,sales_Jan,sales_Feb,sales_March,sales_April]
sales_data = np.array([
[1,44500,40000,49500,47000],  #Paradise_biryani
[2,38500,40000,42500,45000],  #Beijing_bites
[3,64000,58500,56000,61000],  #Pizza_hub
[4,33500,31800,30000,35500],  #Burger_house
[5,77000,79000,72000,73500]   #Chai_point
])
print("===zomato sales analysis===")
print("\n sales_data shape", sales_data.shape)
print("\n sample data for first 3 restau:", sales_data[:3])
```

```python
#Total sales of first 4 month
print("Total sales(INR): \n", np.sum(sales_data[:, 1:]))
# sales of each month
monthly_totals = np.sum(sales_data[:, 1:], axis = 0)
print("Monthly totals: \n", monthly_totals)
```

```python
# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis = 1)
print("Minimum_sales per restaurant: \n", min_sales)
max_sales = np.max(sales_data[:, 1:], axis = 1)
print("Maximum_sales per restaurant: \n", max_sales)
```

```python
# Max sales per month
max_sales1 = np.max(sales_data[:, 1:], axis = 0)
print("Maximum sales per month \n", max_sales1)
```

```python
# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis = 1)
print("Average sales per restaurant \n", avg_sales)
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print("cumsum \n", cumsum)
avg_cumsum = np.mean(cumsum[:, 1:], axis = 1)
print("Average cumulative sales per restaurant \n", avg_cumsum)
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average cumulative sales per month across all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()

```

```python
#vector addition 
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("vector addition: \n ", vector1 + vector2)
print("vector multiplication \n", vector1 * vector2)
print("Dot product \n", np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
